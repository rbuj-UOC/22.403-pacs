# Exercici 1

L'amo d'un heliport ha dissenyat un sistema per a distribuir les arribades dels helicòpters entre les diferents zones del lloc.

Hi ha quatre helicòpters 1, 2, 3 i 4 i dos possibles sentits, ”E” (est) i ”O” (oest).

Les regles de distribució són les següents:

- A l'helicòpter 1 amb sentit "est" li correspon la zona 8 i sentit "oest" la zona 7.
- A l'helicòpter 2 corresponen la zona 10 (est) i 9 (oest).
- A l'helicòpter 3 són 12 i 13 (est i oest respectivament).
- A l'helicòpter 4 li corresponen la zona 14 (est) i 15 (oest).

Dissenyeu la funció zona(*helicopters,sentit)* que rebi com a paràmetres un valor enter positiu que representa l'helicòpter i un caràcter que indiqui el sentit(”E” per a est i ”O” per a oest).

El resultat haurà de ser el número de la zona corresponent. En cas que
l'helicòpter o el sentit no siguin vàlids, ha de retornar un 0.

Com a segona part d'aquest exercici, **utilitzant programació funcional**,
realitza el següent:

- Mostra el resultat de les següents combinacions.
- Mostrar les tuples que es poden realitzar.

<pre>
[ [3,"O"],[7,"O"],[2,"E"] ]
</pre>
<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(1.0 punt)**

```python
import math


def zona(helicopters: int, sentit: str) -> int:
    """Funció que retorna el número de la zona d'un helicòpter.
    La zona depèn del número de l'helicòpter i del sentit de vol (E/O).

    Args:
        helicopters (int): numero de l'helicòpter
        sentit (str): sentit de vol (E/O)

    Returns:
        int: el número de la zona
    """
    answer = math.nan
    try:
        if helicopters not in {1, 2, 3, 4}:
            raise ValueError("helicopters: 1 <= helicopters <= 4")
        if sentit not in {"E", "O"}:
            raise ValueError("sentit: E, O")
        lookup_dict = {
            1: {"E": 8, "O": 7},
            2: {"E": 10, "O": 9},
            3: {"E": 12, "O": 13},
            4: {"E": 14, "O": 15},
        }
        answer = lookup_dict[helicopters][sentit]
    except ValueError as error:
        answer = 0
    return answer


def zone(x) -> list:
    # transposar la matriu
    a, b = zip(*x)
    return list(map(zona, a, b))
```

```python
# 1.1 Mostra el resultat de les següents combinacions
zone([[3, "O"], [7, "O"], [2, "E"]])
```

```python
# 1.2 Mostrar les tuples que es poden realitzar


def check_zone(helicopters: int, sentit: str) -> bool:
    if helicopters not in {1, 2, 3, 4}:
        return False
    if sentit not in {"E", "O"}:
        return False
    return True


def check_zona(x) -> bool:
    return check_zone(x[0], x[1])


list(filter(check_zona, [[3, "O"], [7, "O"], [2, "E"]]))
```

Un inversor ha realitzat diferents transaccions en una cartera d'inversió i ha obtingut bastant benefici al llarg d'un any. Ara, li agradaria canviar tot el benefici que ha obtingut en AED (Dírham dels Unió dels Emirats Àrabs) a una altra moneda.

Per fer-ho, ha decidit fer un canvi de divisa i les divises que ha seleccionat són les següents:

- Euro: 1 AED -> 0,26233 EURO
- Dòlar: 1 AED -> 0,27227 USD
- Ien: 1 AED -> 41,0084 IEN
- Sols: 1 AED -> 1,01084 SOL

Les dades de les transaccions es troben en l'Excel inversion.xlsx.

Per a poder realitzar l'exercici carregui el fitxer inversion.xlsx en un dataframe.

- 2.1: (0,5 punts) Es demana crear una funció "`convertir_divises`" que creï una nova columna per cada divisa en el dataframe on es faci la conversió de la seva moneda (AED) a les divises triades.

- 2.2 (0,5 punts) Es demana crear una funció que s'anomeni: "`total_divises`" i que prengui com a entrada l'arxiu Excel i que retorni el nom de les divises i la suma total generada durant l'any per cadascuna d'elles.

- 2.3 (0,5 punts) Ara, a aquest inversor li agradaria saber si amb els seus guanys obtinguts durant un any podria anar-se'n a viure a un dels països de les divises triades. Per a saber-ho, es demana crear la funció "`pais_mes_accessible`" que tindrà com a entrada el resultat de l'apartat anterior. D'altra banda, el resultat d'aquesta funció serà el país on podria anar a viure. El país triat serà el que tingui major excedent.

Ha de tenir en compte el cost de vida veure quina seria la millor opció

Es pot prendre com a referència els següents costos de vida (mensual):

- Espanya (euro): 1.200 euros
- EUA (dòlar): 5.857 dòlars
- Japó (ien): 425.710 iens
- Perú (sol): 3.927 sols

**Nota:**

Per a realitzar el tercer apartat podeu dividir de manera proporcional la suma total calculada en l'apartat 2 i d'aquesta manera veure mensualment si és factible viure en aquest país.

Recorda mostrar els resultats per pantalla en cada subapartat

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(1.5 punts)**

```python
import pandas as pd


def convertir_divises(filename: str) -> pd.DataFrame:
    """Funció que converteix les divises AED a altres divises.

    Args:
        df (pd.DataFrame): DataFrame amb les dates i les divises AED.

    Returns:
        pd.DataFrame: DataFrame amb les divises convertides.
    """
    # llegir el fitxer XLSX
    df = pd.read_excel(filename)
    # canviar el nom de les columnes
    df = df.rename(columns={"Divisa": "AED"})
    df = df.rename(columns={"Fecha": "Date"})
    # convertir a altres divises
    exchange = {
        'USD': 0.27227,
        'EUR': 0.26233,
        'IEN': 41.0084,
        'SOL': 1.01084
    }
    for currency, rate in exchange.items():
        df[currency] = df['AED'] * rate
    return df


convertir_divises('data/inversion.xlsx')
```

```python
def total_divises(filename: str) -> dict:
    """Funció que retorna el total de les divises EURO, USD, IEN i SOL
    d'un fitxer XLSX.

    Args:
        filename (str): el nom del fitxer XLSX

    Returns:
        dict: un diccionari amb el total de les divises EURO, USD, IEN i SOL
    """
    df = convertir_divises(filename)
    return {
        x: float(df[x].sum()) for x in df.columns if x in ["EUR", "USD",
                                                           "IEN", "SOL"]
    }


total_divises('data/inversion.xlsx')
```

```python
def pais_mes_accessible(divises: dict) -> str:
    """Funció que retorna el país amb ratio més alt.

    Args:
        df (pd.DataFrame): DataFrame amb les divises EUR, USD, IEN i SOL

    Returns:
        str: el nom del país amb índex més alt
    """
    ds = {"EUR": {"surplus": 1200, "country": "Spain"},
          "USD": {"surplus": 5857, "country": "United States"},
          "IEN": {"surplus": 425710, "country": "Japan"},
          "SOL": {"surplus": 3927, "country": "Peru"}}
    dictionary = dict(map(lambda x: (x, divises[x] / ds[x]["surplus"]),
                          ds.keys()))
    dictionary = dict(sorted(dictionary.items(), key=lambda x: x[1],
                             reverse=True))
    currency = list(dictionary.keys())[0]
    return ds[currency]["country"]


print(
    "El país més accessible és {0}.".format(
        pais_mes_accessible(total_divises("data/inversion.xlsx"))
    )
)
```

Es desitja saber si tenim les peces necessàries per a construir una escultura.

Per fer-ho, es tenen els següents diccionaris:

- peces: diccionari que indica les peces que es tenen per a construir l'escultura
- caixa: diccionari que indica les peces que fan falta per a construir l'escultura

Per a saber si tenim les peces necessàries, es demana crear una funció `comptar_peces(peces,caixa)` en la qual s'indiquen els diccionaris i el resultat haurà de ser un diccionari amb les peces que falten per a realitzar l'escultura.

<pre>
peces = {'vermell':10, ’blau’:15, 'groc':25, ’verd’: 10, ’negre’:10}
</pre>

<pre>
caixa = {’gris’: 19, 'vermell':10, ’blau’:30, ’groc’:25,’taronja’: 13}
</pre>

<pre>
comptar_peces(peces,caixa)
{’blava’:15, ’gris’: 19, ’taronja’: 13}
</pre>


<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(1.5 punts)**

```python
def comptar_peces(peces: dict, caixa: dict) -> dict:
    """Funció que compta les peces per fer una escultura

    Args:
        peces (dict): _peces que es tenen per fer l'escultura_
        caixa (dict): _peces que es necessiten per fer l'escultura_

    Returns:
        dict: _peces que falten per fer l'escultura_
    """
    r = {}
    for key, value in caixa.items():
        if key in peces:
            if caixa[key] == peces[key]:
                continue
            r[key] = value - peces[key]
        else:
            r[key] = value
    return r


peces = {'vermell': 10, 'blau': 15, 'groc': 25, 'verd': 10, 'negre': 10}
caixa = {'gris': 19, 'vermell': 10, 'blau': 30, 'groc': 25, 'taronja': 13}
comptar_peces(peces, caixa)
```

# Exercici 4

Es demana realitzar una funció que calculi l'IMC (Índex de Massa Corporal) d'un nombre determinat de persones. Per fer-ho, es té l'Excel anomenat IMC.xslx.

En aquest Excel hi ha la següent informació:

- ID: identificador de les persones
- Nom: nom de les persones
- Edat: edat de les persones
- Altura: altura de les persones (cm)
- Pes: pes de les persones (kg)

La funció que se sol·licita s'anomenarà `imc`, i rebrà:

- L'arxiu de dades en format *.xlsx*.

- Un número comprès entre 1 i 200.

El número comprès entre 1 i 200 és el número que s'utilitzarà per a generar els números aleatoris. Aquests números aleatoris també hauran d'estar compresos entre 1 i 200, ja que 200 és la quantitat de dades que tenim en l'excel.

Els números aleatoris generats correspondran al ID de les persones, i seran els que se seleccionaran.

Finalment, la sortida serà un diccionari on les claus seran els noms de les persones i el seu valor correspondrà a l'IMC calculat.


Exemple:
<pre>
imc(IMC.xslx , 2 )
# aquesta funció generarà 2 números aleatoris,
# en aquest exemple els 2 números aleatoris generats són: 10 i 40
# Si veiem l'ID 10 i 40, veiem que es correspon amb les persones "Amparo" i "Esteban"

# La sortida final serà la següent:

{’Empara’: 19.96, ’Esteban’: 39.88}
</pre>

**Nota:**

La fórmula de l'IMC és el pes (en kg) dividit per l'altura (en metres) al quadrat.

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> **(2.0 punts)**

```python
from numpy import random


def imc(filename: str, num_identifiers: int) -> dict:
    """Funció que retorna el IMC d'una persona a partir del seu pes i alçada.

    Args:
        filename (str): _fitxer amb les dades de les persones_
        number (int): _número de la persona_

    Returns:
        dict: _IMC de les persones seleccionades_
    """
    df = pd.read_excel(filename)
    # canviar el nom de les columnes
    df = df.rename(columns={"Peso": "Weight"})
    df = df.rename(columns={"Altura": "Height"})
    df = df.rename(columns={"Nombre": "Name"})
    # comprovacions
    size = df.shape[0]
    if num_identifiers < 1:
        raise ValueError("num_identifiers: 1 <= num_identifiers")
    if num_identifiers > size:
        raise ValueError(
            "num_identifiers: 1 <= num_identifiers <= {0}".format(size))
    # seleccionar un nombre aleatori d'identificadors
    identifiers = random.randint(len(df), size=(num_identifiers))
    df = df.loc[df['ID'].isin(identifiers)]
    # calcular IMC
    df['IMC'] = df['Weight'] / ((df['Height']/100) ** 2)
    return dict(zip(df['Name'], df['IMC']))


imc('data/IMC.xlsx', 2)
```

# Exercici 5

Per a aquest exercici, es desitja saber quines són les paraules que es repeteixen més d'una vegada en un text, per la qual cosa s'haurà de crear una funció que llegeixi un text i calculi quantes vegades es repeteixen totes les paraules i mostri les paraules que es repeteixen més d'una vegada. L'arxiu a utilitzar és: Elladron.docx.

Es demana crear la funció `lectura` que rebrà la ruta on es troba l'arxiu i haurà de realitzar el següent:

1. Lectura de l'arxiu
2. Comptatge de totes les paraules de l'arxiu
3. Filtrar les paraules que es repeteixin més d'una vegada.

El resultat d'aquesta funció seran les paraules (que es repeteixen més d'una vegada) amb les vegades que es repeteix.

*Notes importants:*

1) Les paraules es compten per si soles, és a dir, si una paraula forma part d'una altra no compta.

2) Si una paraula té alguna lletra majúscula i una altra no, però és la mateixa paraula, es compten com a iguals, per tant, suma al comptador de la paraula.

Exemple:

<pre>
Frase: La comida de la selva es mejor que la lasaña.

Conteo de palabras:

la: 3
comida: 1
de: 1
selva: 1
es: 1
mejor: 1
que: 1
lasaña: 1

Resultado de la función:

la: 3
</pre>

**Nota:**

En tractar-se d'un arxiu .docx ha d'utilitzar la llibreria adequada per a la seva lectura.

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> **(3.0 punts)**

```python
import os

if os.getenv("COLAB_RELEASE_TAG"):
    # instal·la les dependencies
    %pip install python-docx
```

```python
from docx import Document


def lectura(filename: str) -> dict:
    """Funció que llegeix un fitxer DOCX i retorna un diccionari amb les
    paraules i el nombre de vegades que apareixen.
    Args:
        filename (str): el nom del fitxer DOCX
    Returns:
        dict: un diccionari amb les paraules i el nombre de vegades que
        apareixen
    """
    r = {}
    try:
        f = open(filename, "rb")
        document = Document(f)
        for p in document.paragraphs:
            for word in [
                word.strip('.,;:¡!¿?()[]{}"').lower() for word in p.text.split()
            ]:
                if word in r:
                    r[word] += 1
                else:
                    r[word] = 1
        f.close()
        # elimina les paraules que apareixen una sola vegada
        r = dict(filter(lambda x: x[1] > 1, r.items()))
        # ordena el diccionari per nombre de vegades que apareix cada paraula
        r = dict(sorted(r.items(), key=lambda x: x[1], reverse=True))
    except FileNotFoundError:
        print("Error: El fitxer no existeix!")
    return r


lectura("data/Elladron.docx")
```
